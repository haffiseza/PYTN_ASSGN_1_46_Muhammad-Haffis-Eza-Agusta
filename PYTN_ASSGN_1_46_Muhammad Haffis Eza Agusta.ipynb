```python
cat_year=criminal.groupby(["major_category", "year"]).sum()["value"].sort_values(ascending=False) 
```

```python
cat_year.unstack().T #transpose
```

 # Area Pots

```python
fig, ax = plt.subplots(figsize=(20,10))
cat_year.unstack().T.plot(kind="area",ax=ax,stacked=False)

plt.title('total') # add a title to the area plot
plt.ylabel('value') # add y-label
plt.xlabel('year') # add x-label

plt.show()
```

# Histogram

```python
criminal.major_category.value_counts().plot(kind='hist')

plt.title('Major_category ') # add a title to the histogram
plt.ylabel('major_category') # add y-label
plt.xlabel('Value') # add x-label

plt.show()
```

```python
criminal[criminal.year==2016].minor_category.value_counts().plot(kind='hist')

plt.title('minor_category ') # add a title to the histogram
plt.ylabel('minor_category') # add y-label
plt.xlabel('Value') # add x-label

plt.show()
```

# Bar Charts (Dataframe)

```python
fig, ax = plt.subplots(figsize=(20,10))
cat_year.unstack().T.plot(kind="barh",ax=ax,stacked=False)

plt.xlabel('Variables') # add to x-label to the plot
plt.ylabel('year') # add y-label to the plot
plt.title('Major_category') # add title to the plot

plt.show()
```
